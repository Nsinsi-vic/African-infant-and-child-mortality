```python
merged_df1 = pd.merge(data1, data2, on = ["Year", "Entity"], how = "outer")
merged_df2 = pd.merge(merged_df1, data3, on = ["Year", "Entity"], how = "outer")
merged_df3 = pd.merge(merged_df2, data5, on = ["Year", "Entity"], how = "outer")
merged_df4 = pd.merge(merged_df3, data6, on = ["Year", "Entity"], how = "outer")
merged_df5 = pd.merge(merged_df4, data7, on = ["Year", "Entity"], how = "outer")
merged_df5
```

```python
# Calculate year-wise medians and fill missing values
# Under-fifteen mortality rate
merged_df5["Under-fifteen mortality rate"] = merged_df5.groupby("Entity")["Under-fifteen mortality rate"].transform(
    lambda x: x.fillna(x.median())
)
# Death Esimates
merged_df5["Death Esimates"] = merged_df5.groupby("Entity")["Death Esimates"].transform(
    lambda x: x.fillna(x.median())
)
# Under-five mortality rate Deaths per 100 live births
merged_df5["Under-five mortality rate Deaths per 100 live births"] = merged_df5.groupby("Entity")["Under-five mortality rate Deaths per 100 live births"].transform(
    lambda x: x.fillna(x.median())
)
# Estimated maternal deaths
merged_df5["Estimated maternal deaths"] = merged_df5.groupby("Entity")["Estimated maternal deaths"].transform(
    lambda x: x.fillna(x.median())
)
# Births attended by skilled health staff (% of total)
merged_df5["Births attended by skilled health staff (% of total)"] = merged_df5.groupby("Entity")["Births attended by skilled health staff (% of total)"].transform(
    lambda x: x.fillna(x.median())
)
# BCG (% of one-year-olds immunized)
merged_df5["BCG (% of one-year-olds immunized)"] = merged_df5.groupby("Entity")["BCG (% of one-year-olds immunized)"].transform(
    lambda x: x.fillna(x.median())
)
# MCV1 (% of one-year-olds immunized)
merged_df5["MCV1 (% of one-year-olds immunized)"] = merged_df5.groupby("Entity")["MCV1 (% of one-year-olds immunized)"].transform(
    lambda x: x.fillna(x.median())
)
# Pol3 (% of one-year-olds immunized)
merged_df5["Pol3 (% of one-year-olds immunized)"] = merged_df5.groupby("Entity")["Pol3 (% of one-year-olds immunized)"].transform(
    lambda x: x.fillna(x.median())
)
# DTP3 (% of one-year-olds immunized)
merged_df5["DTP3 (% of one-year-olds immunized)"] = merged_df5.groupby("Entity")["DTP3 (% of one-year-olds immunized)"].transform(
    lambda x: x.fillna(x.median())
)
merged_df5
```

```python
# Define a threshold for "low" vaccination coverage (e.g., below 70%)
low_vaccination_threshold = 70

# Filter countries with vaccination rates below the threshold
low_vaccination_df = final_df[
    (final_df["BCG (% of one-year-olds immunized)"] < low_vaccination_threshold) |
    (final_df["DTP3 (% of one-year-olds immunized)"] < low_vaccination_threshold) |
    (final_df["MCV1 (% of one-year-olds immunized)"] < low_vaccination_threshold)
]

# Display countries and years with low vaccination rates
print("Countries with low vaccination rates:")
print(low_vaccination_df[["Entity", "Year", 
                          "BCG (% of one-year-olds immunized)", 
                          "DTP3 (% of one-year-olds immunized)", 
                          "MCV1 (% of one-year-olds immunized)"]])
```

```python
# Calculate vaccination rate changes over the years by country
coverage_change = final_df.groupby("Entity")[[
    "BCG (% of one-year-olds immunized)", 
    "DTP3 (% of one-year-olds immunized)", 
    "MCV1 (% of one-year-olds immunized)"
]].agg(["min", "max"])

# Add a column to show improvement in coverage (max - min)
coverage_change["BCG Change"] = coverage_change["BCG (% of one-year-olds immunized)"]["max"] - coverage_change["BCG (% of one-year-olds immunized)"]["min"]
coverage_change["DTP3 Change"] = coverage_change["DTP3 (% of one-year-olds immunized)"]["max"] - coverage_change["DTP3 (% of one-year-olds immunized)"]["min"]
coverage_change["MCV1 Change"] = coverage_change["MCV1 (% of one-year-olds immunized)"]["max"] - coverage_change["MCV1 (% of one-year-olds immunized)"]["min"]

# Sort by least improvement
coverage_stagnation = coverage_change[["BCG Change", "DTP3 Change", "MCV1 Change"]].sort_values(by=["BCG Change", "DTP3 Change", "MCV1 Change"])

print("Countries with stagnated vaccination rates:")
print(coverage_stagnation)
```
